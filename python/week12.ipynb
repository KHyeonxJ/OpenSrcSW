```python
#실습1: 서울의 일교차
import csv
f= open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f,delimiter=',')
next(data)#헤더 넘기기
max_temp = -999
max_date = ''
min_temp = 999
min_date = ''
max_diff = -999    
max_ddate = ''

for row in data:
    if row[-1]=='':
        row[-1]= -999
    row[-1] = float(row[-1])
    if(max_temp < row[-1]):
        max_date = row[0]
        max_temp = row[-1]
    if row[-2] == '':
        row[-2] = 0 
    row[-2] = float(row[-2])
    if(min_temp > row[-2]):
        min_date= row[0]
        min_temp = row[-2]
    dif = row[-1] - row[-2]
    if(max_diff < dif):
        max_ddate = row[0]
        max_diff = dif

print("일교차가 가장 큰 날: {0:s}".format(max_ddate))
print("최저 기온이 가장 낮은 날 : {0:s}".format(min_date))
print("최고 기온이 가장 높은 날 : {0:s}".format(max_date))  
f.close()
```

```python
#실습2: 지하철 이용객
import csv
f= open('subwayfree.csv','r',encoding = 'utf-8')
data = csv.reader(f,delimiter=',')
header = next(data)
linearr = []
redata = []
for line in data:#호선 배열
    if(line[1] not in linearr):
        linearr.append(line[1])
    redata.append(line)
    
for subway in linearr:
    print("=========="+subway+"==========")
    datalist = []
    for col in redata:
        if (subway == col[1]):
            passenger =int(col[4].replace(",",""))+int(col[5].replace(",",""))
            datalist.append([col[3],passenger])
    datalist.sort(key = lambda x:x[1])#오름차순 정렬
    print("이용객이 가장 많은 역 top 3")
    print("1위: ",datalist[-1])
    print("2위: ",datalist[-2])
    print("3위: ",datalist[-3])
    print(" 이용객이 가장 적은 역 top 3")
    print("1위",datalist[0])
    print("2위",datalist[1])
    print("3위",datalist[2])   
f.close()
```
